## Training the model

In this part we describe the step of training our model. Our model consists of two neural layers, including one hidden layer and one output layer.
### Dataset
The dataset is MNIST downloaded from Lecun's website, with each image of size $28\times28$.
### Structure
The first layer is an affine layer with activation function, and the output layer outputs softmax scores.
### Activation Function
We choose the commonly used RELU function, and because RELU is put right after the affine layer, so we define a unit **affine_relu_forward/backward** just for convenience.
### Backpropagation and Loss
To define the loss of softmax scores, we use cross-entropy loss function as it is often used for multi-class classification, represented as $H(p,q)=-\sum_{x}p(x)\log q(x)$, where $p$ is the true distribution and $q$ is the estimated distribution. Specifically, for backpropagation, we use loss as upstream gradient, and use backward layers in **layers.py** to compute the gradient of $W_1, W_2$ and $b_1, b_2$.
### Learning Rate Decay
In this assignment, we set learning rate decay as 0.95, meaning that we time learning rate with 0.95 per epoch.
### $L_2$ Regularization
For the convenience of our computation, we add the regularization as $\dfrac{\lambda}{2}\|W_1\|_F^2$ and $\dfrac{\lambda}{2}\|W_2\|_F^2$, where $\lambda$ is the regularization strength.
### Optimizer
In **optim.py**, we implemented both sgd and adam, after several trials, we find that adam converges faster than sgd, so we use adam as optimizer instead of sgd in this assignment.

Import some necessary packages.

```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

This part is to split the training data into training and validation, and validation data has a portion of 0.3, and the data shape are shown in the output clearly.

```python
from data_utils import load_data, split_train_val

X_train_val, y_train_val, X_test, y_test = load_data('./dataset')
X_train, y_train, X_val, y_val = split_train_val(X_train_val, y_train_val, 0.3)
data = {'X_train':X_train, 'y_train':y_train, 'X_val':X_val, 'y_val':y_val, 'X_test':X_test, 'y_test':y_test}

# show data shape
for k, v in list(data.items()):
    print(('%s: ' % k, v.shape))
```

For this part, we define the size of hidden layer as 512, and learning rate as 2e-4, and train our first model, the loss history is shown as follows.

```python
from project_tools.fc_net import TwoLayerNet
from project_tools.solver import Solver

input_size = 28 * 28
hidden_size = 512
num_classes = 10
model = TwoLayerNet(input_size, hidden_size, num_classes)
solver = Solver(model, data, update_rule='adam', optim_config={'learning_rate':2e-4},
                lr_decay=0.95, num_epochs=10, batch_size=128, print_every=500)
solver.train()
print("Best val acc: ", solver.best_val_acc)
```

This part visualize the loss history and accuracy history.

```python
plt.subplot(2, 1, 1)
plt.title('Training loss')
plt.plot(solver.loss_history, 'o')
plt.xlabel('Iterations')
plt.subplot(2, 1, 2)
plt.title('Accuracy')
plt.plot(solver.train_acc_history, '-o', label='train')
plt.plot(solver.val_acc_history, '-o', label='val')
plt.plot([0.95] * len(solver.val_acc_history), 'k--')
plt.xlabel('Epochs')
plt.legend(loc='lower right')
plt.gcf().set_size_inches(15, 12)
plt.show()
```

## Tuning Hyperparameters

After several trials, we use grid search to search for the best combination of hyperparameter. The criterion is the accuracy on the validation set, and we store our best model in a variable.

```python
best_val_accuracy = 0.0
best_model = None
learning_rate_list = [2e-4, 5e-4, 1e-3]
hsize_list = [400, 512]
reg_list = [0.0, 0.1]
for lr in learning_rate_list:
    for hsize in hsize_list:
        for reg in reg_list:
            model = TwoLayerNet(input_size, hsize, num_classes, reg=reg)
            solver = Solver(model, data, update_rule='adam', optim_config={'learning_rate':lr}, lr_decay=0.95, num_epochs=10, batch_size=128, verbose=False)
            solver.train()
            print('learning rate: ',lr,'hidden_dim: ',hsize,'reg: ',reg)
            print('validation accuracy: ', solver.best_val_acc)
            if solver.best_val_acc > best_val_accuracy:
                best_val_accuracy = solver.best_val_acc
                best_model = model
print('best validation accuracy: ', best_val_accuracy)
```

For our best model, we test it on the test set.

```python
y_test_pred = np.argmax(best_model.loss(data['X_test']), axis=1)
print('Test set accuracy: ', (y_test_pred == data['y_test']).mean())
```

We store our best model as a binary file under the same directory.

```python
file = open('best_model.obj','wb')
pickle.dump(best_model, file)
file.close()
```
